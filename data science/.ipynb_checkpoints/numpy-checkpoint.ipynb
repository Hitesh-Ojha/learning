```python
import time 
```

```python
z = range(10000000)
y = range(10000000)

start_time = time.time()

c = [(z,y) for z,y in zip(z,y)]

print(time.time() - start_time)
```

```python
a = np.arange(10000000)
b = np.arange(10000000 , 20000000)

start_time = time.time()

c = a + b

print(time.time() - start_time )
```

first proves that the space in normal list is higher basicly space complexity is lowwer and other proves that time complexity is lowwer in np array
