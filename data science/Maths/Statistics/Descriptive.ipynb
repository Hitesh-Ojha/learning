```python
# Cumulative Frequency  
```

```python
d = data['Pclass'].value_counts()
```

```python
b = ((d/d.sum())*100).values
```

```python
b
```

```python
c = []
cf = 0 
for i in b:
    cf = cf + i 
    c.append(cf)
```

```python
plt.plot(c , marker = 'D')
```
