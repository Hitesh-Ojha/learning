```python
with open('Data set/plot_data/csv_data_plot.csv') as csv_file: # here use of open to access file and storing itin to variable name cvs_file
    csv_reader = csv.DictReader(csv_file) # by importing Csv , using Dirct Reader we can extract the data into dictionary (key and value bro simple as that)

    language_counter = Counter() #basicly we are creating object of Counter which give us meathod like most_common and storing them into variable
    

    for row in csv_reader: 
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
    # row = next(csv_reader) use for reading first line of data 
    # print(row['LanguagesWorkedWith'].split(';')) we are retriving value with use of key as key = LanguagesWorkedWith and values = python , js , c++ 

languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

for li in languages: 
    print(li) 
print('\n') 
for li2 in popularity: 
    print(li2)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(languages , popularity )
plt.title("most popular languages")
plt.xlabel("programing_languages ")
plt.ylabel("Number of people who used")
plt.xticks(rotation=45, ha='right') # rotation the label 
plt.tight_layout()
#print(languages)
#print(papularity)
#print(language_counter.most_common(15))
```

**Filling Area on Line Plots**

```python
data = pd.read_csv('Data set/plot_data/data_dev_salary.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

plt.fill_between(ages, py_salaries, dev_salaries ,
                  where=(py_salaries > dev_salaries),interpolate = True, color='green' ,alpha=0.25 , label = 'profit margin')

plt.fill_between(ages, py_salaries, dev_salaries ,
                  where=(py_salaries <= dev_salaries),interpolate = True, color ='red',alpha=0.25 , label = 'loss margin')

plt.legend()
plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.tight_layout()

plt.show()
```
