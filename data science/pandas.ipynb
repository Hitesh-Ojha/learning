# lets create a function of this 

```python
def total_match(team_name):
    team1_data = data['team1'].value_counts()
    team2_data = data['team2'].value_counts() 

    played_as_t1 = team1_data[team_name]
    played_as_t2 = team2_data[team_name]

    return played_as_t1 + played_as_t2 
    
```

```python
total_match('Mumbai Indians')
```

```python
def max_run(name):
    mask  = data2['batsman'] == name
    played = data2[mask]
    return played.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending = False).head(1)
```

```python
max_run('G Gambhir')
```

```python
max_run('SK Raina')
```
